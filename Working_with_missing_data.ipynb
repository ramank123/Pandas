```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(
    np.random.randn(5, 3),
    index=["a", "c", "e", "f", "h"],
    columns=["one", "two", "three"],
)


df["four"] = "bar"

df["five"] = df["one"] > 0
```

```python
df
```

```python
df2 = df.reindex(["a", "b", "c", "d", "e", "f", "g", "h"])
```

```python
df2
```

```python
df2["one"]
```

```python
pd.isna(df2["one"])
```

```python
df2["four"].notna()
```

```python
df2.isna()
```
